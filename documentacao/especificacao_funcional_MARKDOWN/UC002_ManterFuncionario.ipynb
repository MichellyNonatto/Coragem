### Descrição Resumida do Caso de Uso
<br>
<small> Este caso de uso tem como objetivo descrever as funcionalidades pertinentes ao Manter Funcionário, tal como, inserir, alterar ou remover funcionário. </small>

| | |
| :------- | :------ |
| **Atores** | Gerente |  
| **Pré-Condição** | O gerente deve ser autenticado no caso `UC001`. |  
| **Fluxo Principal** | **P01.** Na `Interface005` o gerente deve clicar em `"Funcionários"`, sendo direcionado para a `Interface006`. <br> **P02.** O sistema deve retornar uma lista dos funcionarios já cadastrados com algumas informações. <br> **P03.** O gerente clica no botão `"+ novo funcionário"` para realizar o cadastro de um novo funcionário. <br> **P04.** O sistema retorna a `Interface007`, onde o gerente preenche as informações do novo funcionário. <br> **P05.** O gerente clica no botão `Cadastrar funcionário`, e o sistema realiza o cadastro do funcionário. <br> **P06.** O sistema retorna para a `Interface008` de detalhes do novo funcionário cadastrado. |
| **Fluxo Alternativos** | **A01 [P02] - Alterar Funcionário** <br> 1. O gerente clica em `Editar` na linha do funcionário a ser alterado. <br> 2. O sistema direciona para a `Interface008`, o gerente altera as informações necessárias. <br> 4. O gerente clica em `"Editar funcionário"`. <br> 5. O sistema salva as alterações no banco de dados e retorna a `Interface008`. <br> **A02 [P02] - Excluir Funcionário** <br> 1. O gerente clica em `Excluir` na linha do funcionário a ser excluido. <br> 2. O sistema direciona para a `Interface010`, o gerente seleciona `"Excluir"`. <br> 4. O sistema retorna a `Interface006`. <br> **A03 [P02] - Ver Funcionário** <br> 1. O gerente clica em `Ver funcionário` na linha do funcionário a ser exibido, indo para a `Interface008`. <br> 2. O gerente pode clicar em `"Editar funcionário"` que realizar o fluxo **A01.2** ou em `"Excluir Funcionário"` que realiza o fluxo **A02.2** | 
| **Exceções** | **E01 [A01.2] - Informação incompleta** <br> 1. O sistema exibe a mensagem MSG003. <br> 2. O sistema retorna a Interface007. <br> **E02 [A01.3] - Funcionário já existente** <br> 1. O sistema exibe a mensagem MSG004. <br> 2. O sistema retorna a Interface006. <br> | 
| **Pós-Condição** | **PC01 [P05]** <br> 1. O sistema exibe a mensagem `MSG001`. <br> **PC02 [A01.5]** <br> 1. O sistema exibe a mensagem `MSG002`. <br> **PC03 [A02.2]** <br> 1. O sistema exibe a mensagem `MSG003`.| 
| **Produtos Gerados** | Não se aplica. | 
| **Requisitos** | O sistema deve cadastrar os funcionários e suas funções: Foto de perfil, Nome Completo, Número de telefone, Logradouro, CEP, Documento (CPF, RG ou CNH), E-mail e Função. | 
| **Regras de Negócio** | **RN01** Para todos os funcionários deve ser criado um e-mail comercial da empresa. <br> **RN02** O sistema deve deletar o e-mail comercial do funcionário assim que o desligamento do mesmo for efetuado. <br> | 
| **Casos de Uso Incluídos** | Não se aplica. | 
| **Casos de Uso Estendidos** | Não se aplica. |

| MSG002 | Alteração em funcionário salva com sucesso! |
| MSG003 | Funcionário excluido com sucessso! |
