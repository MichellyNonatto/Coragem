### Descrição Resumida do Caso de Uso
<small> Esse caso de uso tem como objetivo executar as atividades de criar e pesquisar cliente. </small>

| | |
| :------- | :------ |
| **Atores** | Funcionário|  
| **Pré-Condição** | O funcionario deve estar em processo de criação de Pet do caso `UC004`.|  
| **Fluxo Principal** | **P01.** O funcionário, na `Interface0014` irá clicar em `"+ Adicionar pet"` e será direcionado a `Interface015`. <br> **P02.** O sistema retorna uma lista com informações dos clientes já cadastrados no sistema. <br> **P03.** O funcionário clica em `"Ver Tutor"` sendo direcionado a `Interface012` do caso de uso `UC003`, retornando informações do tutor. <br> **P04.** O funcionário pode editar as informações do tutor clicando em `"Editar tutor"`, sendo direcionado a `Interface019`. <br> **P06.** O funcionário pode deletar o tutor clicando em `"Excluir tutor"`, direcionando para a `Interface013`. <br> **P07.** O funcionário pode cadastrar um novo tutor cliando em `"+ Adicionar tutor"`, direcionando a `Interface018` <br>  **P08.** O funcionário clica em `"Vincular pet"` direcionando a `Interface016` do caso de uso `UC004`. <br> **P09.** Fim do caso.  |
| **Fluxo Alternativos** | **A01 [P02] - Pesquisar Cliente** <br> 1. O funcionário poderá pesquisar o cliente a partir do nome, número ou pelo e-mail cadastrado. <br> 2. O sistema retorna informações conforme o solicitado. <br> **A02 [P02] - Criar novo cliente** <br> 1. Na interface `Interface015`, o funcionário clica em `"+ Adicionar Tutor"`. <br> 3. O sistema redireciona a `Interface018`. <br> 4. O funcionário preenche os campos conforme for solicitado. <br> 5. O funcionário clica em `"Cadastrar tutor"`. <br> 5. O sistema registra o novo cliente e redireciona para a `Interface015`. <br>**A03 [P04] - Editar Tutor** 1.Na interface `Interface012`, o funcionário clica em `"Editar Tutor"` sendo direcionado para `Interface019`. <br> 2. O funcionário modifica as informações necessárias. <br> 3. O sistema retorna a mensagem `MSG001` e retorna a `Interface012`. | 
| **Pós-Condição** |Será pesquisado ou criado na instância cliente.| 
| **Produtos Gerados** | Não se aplica. | 
| **Requisitos** | O sistema deve registrar informações dos tutores: Nome Completo, Número de telefone, Logradouro, CEP, Documento (CPF ou RG), E-mail. <br> O sistema deve enviar relatórios de comportamento simples para o tutor via e-mail. 
| **Regras de Negócio** |  1. O tutor é a pessoa contratante do serviço.| 
| **Casos de Uso Incluídos** | Não se aplica |
