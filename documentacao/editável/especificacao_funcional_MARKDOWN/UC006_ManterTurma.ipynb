### Descrição Resumida do Caso de Uso
<small> Esse caso tem como objetivo fazer atividade de CRUD nos serviços e nas turmas.</small>

| | |
| :------- | :------ |
| **Atores** | Funcionário|  
| **Pré-Condição** | Deve haver a existencia de um pet e um funcionário responsável pelo cuidado, cadastrados na base de dados. |  
| **Fluxo Principal** | **P01.** O funcionário clica em `"Turmas"`. <br> **P02.** O sistema direciona para a `Interface020` com uma lista de turma cadastradas. <br> **P03.** O funcionário seleciona a turma desejada. <br> **P07.** Fim do caso `UC006`. |
| **Fluxo Alternativos** | **A01 [P01] - Editar Serviço**<br>1. O funcionário clica em `"Serviços"` sendo direcionado para a `Interface021`. <br> 2. O sistema retorna uma lista de serviços cadastrados no sistema, nome, dia da semana que é realizado o serviço e valor. <br> 3. O funcionário clica em `"Ver serviço"` no serviço desejado. <br> 4. O sistema direciona para a `Interface022` . <br> 5. O funcionário pode editar o serviço em `"Editar serviço"`, sendo direcionado para a `Interface023`.<br>6. O funcionário salva as modificações sendo direcionado a `Interface022` com a mensagem `MSG001`. <br> **A02 [A01.2] - Adicionar Serviço** <br> 1. O funcionário pode adicionar o serviço clicando em `"+ Adicionar serviço"`, sendo direcionado para a `Interface023`. <br> 2. O funcionário salva a nova instância de serviços e é direcionado a `Interface022` com a mensagem `MSG002`. <br>**A03 [P02] - Pesquisar Turmas** <br> 1. O funcionário pode perquisar a turma pelo seu nome. <br> 2. O sistema retorna as turmas existentes com base na pesquisa feita pelo funcionário. <br> **A04 [P02] - Adicionar Turma** <br> 1. O funcionário clica em `"+ Adicionar turma"`, sendo direcionado para a `Interface025`. <br> 2. O funcionário preenche os serviços que serão fornecidos. <br> 3. O funcionário salva a nova instância de turma em `"Adicionar turma"`, sendo direcionado para a `Interface020` com a mensagem `MSG003`.<br> **A05 [P03] - Editar Turma** <br> 1. O funcionário clica em `"Ver turma"` sendo direcionado para a `Interface025`. <br> 2. O funcionário clica em `"Editar turma"` e é direcionado para a `Interface027`. <br> 3. O funcionário modifica o nome e os serviços vinculados a turma e clica em `"Editar turma"`. <br> 3. O sistema redireciona para a `Interface020` com a mensagem `MSG004`. | 
| **Exceções** | Não se aplica. |

| **Regras de Negócio** | Cada turma pode conter de 1 a n serviços cadastrados. | 
| **Casos de Uso Incluídos** | Não se aplica.  |
